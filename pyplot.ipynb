## Basic plots using matplotlib.pyplot
***

Let's start with a very basic linear plot:

```python
x = np.array([0,2,4,6,8,10])
y = np.array([10,20,30,40,50,60])

plt.plot(x, y)
plt.show()
```

What we just did was to define 2 arrays (x and y) with the same lenght and plot one against the other using the function `plt.plot()`. All other parameters were set to default.`plt.plot()` accepts three basic parameters:<br><br>
- ***x,y***: arrays to be plotted
- ***fmt***: formatting string, as:
  -`fmt = '[marker][line][colour]`. For example, 'r-' will render a red solid line. A complete set of fmt strings is available in the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).
- ***data***: Provides a label for indexed data to be plotted.

### Markers

```python
#Plotting as red circles
plt.plot(x, y,'ro')
plt.show()
```

<br>

`plt.plot()` also allows us to plot multiple points and format how the marks should look like. In the plot below, the marks are stars.

```python
points = np.array([7, 8, 6.5, 9.5])

plt.ylabel('Points')
plt.plot(points, marker = '*')
plt.show()
```

### Labels

Now let's add some labels to the first plot:

```python
#Adding labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, y)
plt.show()
```


<br>Note that when no value is provided for x, `plt.plot()` will plot the values for y and assume `x = range(len(y))`, starting from 0.

```python
#Plotting only y
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(y)
plt.show()
```

<br>Now look at what happens to the y-axis when we plot only x:

```python
#Plotting only x
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x)
plt.show()
```

In the plot above, x was assumed as the value to be plotted on the y-axis. This happens because pyplot assumes x as an optional value - if you only pass one parameter, it will take it as the value for y.

### Plotting multiple subplots

We can create multiple subplots by using `plt.subplot()`, which take 3 parameters: (rows,columns,index).<br>

```python
#first subplot:
year = np.array([1990, 2000, 2010, 2020])
house_price = np.array([20, 40, 100, 200])
plt.ylabel('Thousand US$')
plt.subplot(1, 2, 1) #1 row, 2 columns, position 1
plt.plot(year,house_price)
plt.title("House prices")

#second subplot:
salary = np.array([70,50,45, 40])
plt.ylabel('Thousand US$')
plt.subplot(1, 2, 2) #1 row, 2 columns, postion 2
plt.plot(year,salary)
plt.title("Annual salaries")

plt.tight_layout(pad=3) #Add space between subplots
plt.show()
```

## Histograms
***

Pyplot also allows us to use other types of plots, such as bar charts, histograms, scatterplots, and pie charts. I this section we'll look at how to plot a histogram using `plt.hist()`.

Histograms are approximated visual representations of numerical distributions. For instance, let's say we have the final grades of all 70 students in a class. How could we represent it?

First, let's create an array with 70 random grades, with a [truncated normal distribution](https://en.wikipedia.org/wiki/Truncated_normal_distribution) centred in 5 and with a standard deviation of 2. For this we'll use Scipy's function `truncnorm()`, which is part of the `stats` module that we have previously imported.

```python
lower, upper = 0, 10 #define lower and upper boundaries
mu, sigma = 6, 2     #define centre and std deviation
grades = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
```

Now we plot this using `plt.hist()`

```python
plt.xlabel('Grade')
plt.ylabel('Number of students')
plt.title('Distribution of grades')
plt.hist(grades.rvs(70)) #rvs to define sample size
plt.show()
```

## Styling your plots
***

As you might have noticed by now, basically every aspect of your plots can be customised. For instance, we can customise the looks of our plots by defining parameters for colours, number of bins, density etc. For a comprehensive list of parameters, check the `plt.hist()` [official documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html).

```python
#Create dictionaries with parameters for title and axes fonts
title = {'fontname':'Comic Sans MS','fontsize':14}
axes = {'fontname':'Arial','fontsize':10}

#Add parameters to xlabel, ylabel, and title
plt.xlabel('Grade',axes)
plt.ylabel('Number of students',axes)
plt.title('Distribution of grades',title)

#Change color of bins and edges
plt.hist(grades.rvs(70), edgecolor='black',color='green')
plt.show()
```

### Plot 1: Scatterplot

Scatterplots are visual representations useful to see if there is any correlation between two variables.

Let's exemplify this using the famous [Iris dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set), which has measurements of petal and sepal width and length for three species of Iris flowers - setosa, virginica, and versicolor. Can we differentiate these three species by their petal or sepal sizes?

We start by importing `pandas`, so we can manipulate the dataset taken from the internet.

```python
#Import pandas
import pandas as pd

#Dataset source
urliris='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#Read dataset as a pandas dataframe
iris = pd.read_csv(urliris,names=['sepal_length', 'sepal_width', 'petal_length','petal_width', 'species'])

#show a sample of 10 random rows
iris.sample(10)
```

<br>However, instead of using the `plt.scatter` method, we will use `fig,ax=plt.subplots()`, through which we define 2 variables as subplots: `fig` for the figure, and `ax` for the actual axes where we will plot our values. This provides a more efficient way of handling multivariable plots.

```python
#Define subplots
fig, ax = plt.subplots()

#Select species
setosa = iris[iris['species'] == 'Iris-setosa']
versic = iris[iris['species'] == 'Iris-versicolor']
virgin = iris[iris['species'] == 'Iris-virginica']

#Sepal sizes by species
setosa_sepal = ax.scatter(setosa['sepal_length'], setosa['sepal_width'], label='Setosa')
versic_sepal = ax.scatter(versic['sepal_length'], versic['sepal_width'], label='Versicolor')
virgin_sepal = ax.scatter(virgin['sepal_length'], virgin['sepal_width'], label='Virginica')

#Set labels and title
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_title('Sepal sizes')

#Add a legend
ax.legend()

plt.show()
```

As we can see, setosas's sepals are fairly distinguishable in size from the other two species. Versicolor and virginica's points, on the other hand, are all mixed up, maybe with a tendency for virginica's sepals to be slightly larger.

So, sepal sizes are no good indication of which species we are talking about. Let's try their petals, then.

```python
fig, ax = plt.subplots()

#Petal sizes by species
setosa_petal = ax.scatter(setosa['petal_length'], setosa['petal_width'], label='Setosa')
versic_petal = ax.scatter(versic['petal_length'], versic['petal_width'], label='Versicolor')
virgin_petal = ax.scatter(virgin['petal_length'], virgin['petal_width'], label='Virginica')

ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
ax.set_title('Petal sizes')
ax.legend()

plt.show()
```

Still not perfect, but seems like petal sizes are a much better indicator than sepal sizes. Setosas' petals are clearly much smaller than the other 2 species, with all the points clustered arount 2(length) and 1 (width). Virginica's petals are mostly larger that versicolor's, but there are still a couple of mixed up points, so we can't say that virginica's petals are *always* larger than versicolor's.

Scatterplots are a quick and useful method of visualising and getting insights from data. In our case, it showed that petals are greater indicators of Iris species than sepals.

### Plot 2: Piechart


Pyplot also allows us to plot piecharts, using the `pie()` function. Pie charts provide a quick glance of proportions betweens different parts of a whole. For instance, we can draw a pie chart to have an idea of how much time we spent on each activity of ~~this assignment~~ a fictitious project.
<br>We start by defining our variables - in this case, the number of hours - as an array of numerical values, which will determine the relative size of the our chart's **wedges**, or its component pieces. We then set our labels, colours and explode (i.e. how far from the centre each wedge is - here, we only highlighted the last wedge).

Again, here we'll use the `fig,ax = subplots()` method, setting the size of our figure to 10. Note that by default the first wedge starts just above the horizontal axis (x-axis) and subsequant wedges are 'stacked' counterclockwise. We can change the starting point by defining a `startangle` in degrees. In our case, we changed it to 180.

We are interested in the proportions here, so instead of plotting the number of hours we decided to show the percentage of time taken by each activity. To do that, we give `autopct` a string value of '%.1%%', meaning 1 decimal point and showing the '%' character after. If we wanted 2 decimal points, we could simply change it to '%.2%%'.

```python
#Set variables
#Values, labels, palette, and explode must be the same size
hours = np.array([30,20,15,5,5])
activity = ['Procrastinating','Panicking','Researching','Writing a README','Coding']
colours = ['#072a42','#272a42','#472a42','#672a42','#872a42']
slicing = np.array([0,0,0,0,0.3]) #Distance of each wedge from centre

#Define size of figure
fig, ax = plt.subplots(figsize=(10,10))

#Title
ax.set_title('Time spent on project', fontsize = 18)

#Pie chart plot
ax.pie(
       hours, labels = activity, colors = colours,
       startangle=180, 
       explode = slicing, 
       shadow = True,         #Add shadow to wedges
       autopct='%.1f%%'      #Show percentages with 1 decimal
      )

#Add legend
ax.legend()

#Remove tilting
plt.axis('equal')

plt.show() 
```

### Plot 3: Boxplot

Boxplots show the distribution of data composed of five summary numbers:
1. **Minimum:** the smallest value within a range in the data.
2. **First quartile (Q1 or 25th percentile):** the middle number between the smallest number (not the 'minimum') and the median of the dataset.
3. **Median (Q2 or 50th Percentile):** the middle value of the dataset.
4. **Third quartile (Q3 or 75th percentile):** the middle value between the median and the highest value (not the 'maximum') of the dataset.
5. **Maximum:** the largest value within a range in the data.

This is how a basic boxplot looks like:

    <figcaption><strong>Elements of a boxplot</strong>. Galarnyk (2018), towardsdatascience.com</figcaption>
</figure>


In the image above, the minimum and maximum values are displayed as the ends of the blue **whiskers**, whilst boundaries of the 'box' represent Q1 and Q3, with the space between these 2 values being the **Interquartile range (IQR)**. Also, notice that outside the minimum to maximum range there are still some **outliers**. These are values that are not within a normal distance from the rest of the data (i.e. whatever is smaller than Q1 - 1.5\*IQR or greater than Q3 + 1.5\*IQR).

To exemplify the use of boxplots, let's go back to our iris dataset.

First let's try to plot all the different parameters (sepal and petal sizes) irrespective of the species and see what we get from there.

For this, we'll first have to skip the 'species' column from our analysis (its values are strings, so it wouldn't be plotted anyway). However, instead of simply removing that column, we create a new dataframe with species being our index, like this:

```python
#Index by species
iris_species = iris.set_index('species')

#Show a sample
iris_species.sample(10)
```

```python
#Plot using plt.boxplot()
fig,ax = plt.subplots(figsize=(10,10))
labels = ['Sepal widths', 'Sepal lengths', 'Petal widths', 'Petal lengths']

ax.boxplot(iris_species, labels = labels)
plt.show()
```

For an exploratory analysis of the dataset, we can say it its 'okay-ish'. It shows us that most sepal widths measure between 5.1 and 6.5cm, most sepal lengths between 2.8 and 3.3cm, petal widths between 1.6 and 5.1cm, and petal lengths between 0.4 and 1.8cm. There are some outliers for sepal lengths, we get the median, minimum and maximum values... but that's it. There's not much more information this plot can provide us with.

So we need a different approach to get more insightful information from the iris dataset. We need, for instance, to see how each of these parameters vary between the 3 species. That's how we do it:

For the sake of simplicity, we'll shorten the names of the species.

```python
#Replace species with simpler names
iris['species'].replace(['Iris-setosa','Iris-versicolor','Iris-virginica'],['Setosa','Versicolor','Virginica'], inplace=True)

iris.sample(10)
```

Then, we just need to define our variables, our labels (the three species, which are the unique values in the column 'species'), and our plots. For each parameter, we define a list containing all values of each of the columns, separated by species).

Our plotting method will be the `fig,ax = subplots()` again, but notice that now we have 4 plots, placed in 2 rows and 2 columns. This allows us to manipulate each of the plots easily, but in our case we just added a title for each. 

```python
#Sepal lengths
setosa_sl = iris[iris['species'] == 'Setosa']['sepal_length']
versic_sl = iris[iris['species'] == 'Versicolor']['sepal_length']
virgin_sl = iris[iris['species'] == 'Virginica']['sepal_length']
sl = [setosa_sl, versic_sl, virgin_sl]

#Sepal widths
setosa_sw = iris[iris['species'] == 'Setosa']['sepal_width']
versic_sw = iris[iris['species'] == 'Versicolor']['sepal_width']
virgin_sw = iris[iris['species'] == 'Virginica']['sepal_width']
sw = [setosa_sw, versic_sw, virgin_sw]


#Petal lengths
setosa_pl = iris[iris['species'] == 'Setosa']['petal_length']
versic_pl = iris[iris['species'] == 'Versicolor']['petal_length']
virgin_pl = iris[iris['species'] == 'Virginica']['petal_length']
pl = [setosa_pl, versic_pl, virgin_pl]

#Petal widths
setosa_pw = iris[iris['species'] == 'Setosa']['petal_width']
versic_pw = iris[iris['species'] == 'Versicolor']['petal_width']
virgin_pw = iris[iris['species'] == 'Virginica']['petal_width']
pw = [setosa_pw, versic_pw, virgin_pw]

#Unique values from column species as labels
labels = iris['species'].unique()


#1 figure, 4 sets of axes, 2 rows, 2 columns
fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2, 2, figsize=(10, 10))

#Set plot titles
ax1.set_title('Sepal lengths')
ax2.set_title('Sepal widths')
ax3.set_title('Petal lengths')
ax4.set_title('Petal widths')


#What to plot in each ax
ax1.boxplot(sl, labels=labels)
ax2.boxplot(sw, labels=labels)
ax3.boxplot(pl, labels=labels)
ax4.boxplot(pw, labels=labels)


plt.show()
```
